````python
##################
## Part 1 of 2
##################
#
# Show your work below. Submit your answers on Canvas.
# You may create additional cells to segment your work.

## Read the data in as a pandas DataFrame
df = pd.read_csv('quiz_2_pca.csv', header=None)
````

````python
# Print dimensionality
print('Dimensionality:', df.shape[1])

# Apply PCA
pca = PCA(n_components=df.shape[1])
pca_data = pca.fit_transform(df)

# Fraction of variance explained
print('first PC dimension:', pca.explained_variance_ratio_[0])
print('last PC dimension:', pca.explained_variance_ratio_[-1])

# Scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(pca_data[:, 0], pca_data[:, 1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scatter plot of PC1 vs PC2')
plt.show()
````

````python
# Number of clusters
num_clusters = 3

# Apply K-means clustering
kmeans = KMeans(3).fit(df_kmeans)
labels = kmeans.predict(df_kmeans)
````
